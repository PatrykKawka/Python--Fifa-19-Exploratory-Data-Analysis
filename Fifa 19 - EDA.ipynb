```python
english_players_numeric = english_players.select_dtypes(include=['int64', 'float64'])
english_players_numeric = english_players_numeric.drop(columns=['ID'])

plt.figure(figsize=(8,6))

sns.heatmap(english_players_numeric.corr(),fmt=".2f",annot=True, annot_kws={"size":10}, cmap = 'plasma', linewidths=0.6, linecolor='black')
plt.title('Matryca korelacji')
plt.show()
```

```python
english_players_club = english_players.groupby('Club').agg({
    'ID': 'count',
    'Age': 'mean',
    'Overall': 'mean',
    'Wage': ['min', 'max', 'median', 'mean', 'std'],
    'Value': ['min', 'max', 'median', 'mean', 'std']
})


english_players_club.columns = ['_'.join(col).strip() if col[1] else col[0] for col in english_players_club.columns.values]

english_players_club = english_players_club.sort_values(by='ID_count', ascending=False)

english_players_club.query('ID_count>11').head(15)
```

```python
#TOP 10
english_players[['Name','Age','Club','Value']].sort_values(by='Value', ascending=False).head(10)
```

```python
ManUnited_players = english_players.query('Club =="Manchester United"')
ManUnited_players.sort_values(by='Overall', ascending=False)
```

```python
UnitedorCity_players = english_players.query('Club == "Manchester United" or Club == "Manchester City"')
UnitedorCity_players.sort_values(by='Wage', ascending=False)
```

```python
plt.figure(figsize=(4, 6))
sns.boxplot(UnitedorCity_players,x='Club', y='Wage')

plt.axhline(UnitedorCity_players['Wage'].median(), color='red', linestyle='--', linewidth=2, label='Median') #linia mediany
plt.axhline(UnitedorCity_players['Wage'].min(), color='blue', linestyle=':', linewidth=2, label='Min') #linia min
plt.axhline(UnitedorCity_players['Wage'].max(), color='green', linestyle=':', linewidth=2, label='Max') # linia max

plt.title('Wykres skrzynkowy pokazujący informacje na temat rozkładu zmiennej wynagrodzenie zawodnika (Wage)')
plt.legend()
plt.show()
```

```python
UnitedorCity_players.groupby('Club').agg({
    'ID': 'count',
    'Age': 'mean',
    'Overall': 'mean',
    'Wage': ['min', 'max', 'median', 'mean', 'std', 'sum'],
    'Value': ['min', 'max', 'median', 'mean', 'std','sum']
    })
```

```python
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

sns.barplot(ax=axs[0], data=UnitedorCity_players, x='Club', y='Value', color='Green')
axs[0].set_title('Wykres przedstawiający wartości angielskich zawodników w klubach')

sns.barplot(ax=axs[1], data=UnitedorCity_players, x='Club', y='Value', hue='Wage', palette='Greens')
axs[1].set_title('Wykres przedstawiający wartości angielskich zawodników w klubach z podziałem na wynagrodzenia')

sns.barplot(ax=axs[2], data=UnitedorCity_players, x='Club', y='Wage', color='Green')
axs[2].set_title('Wykres przedstawiający zarobki angielskich zawodników w klubach')

plt.tight_layout()
plt.show()
```
